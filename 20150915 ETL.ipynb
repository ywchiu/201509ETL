```python
import urlparse
import requests
# get the session
rs = requests.session()
res = rs.post('http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance', data=payload)

# get the page number
rec_number = soup.select('.T11b')[0].text
from math import ceil
page_number =  int(ceil(float(rec_number) / 100))

# open a file
f = open('bid_list.txt', 'w')

# format
page_format  = 'http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance&searchTarget=ATM&method=search&isSpdt=&pageIndex={}'

# page iteration
for page in range(1, page_number+1):
    pageurl = page_format.format(page)
    bid_response = rs.get(pageurl)
    bid_soup = BeautifulSoup(bid_response.text) 
    bid_table = bid_soup.select("#print_area")[0]
    bid_rows = bid_table.select('tr')[1:-1] 
    for bid_row in bid_rows:         
        link_href =  urlparse.urljoin(bid_response.url , \
                           bid_row.select('a')[0]['href'])
        f.write(link_href + '\n')
f.close()    
```
