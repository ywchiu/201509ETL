```python
import requests
res = requests.get('http://www.appledaily.com.tw/realtimenews/section/new/')
print res
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(res.text)
ary=[] 
for li in soup.select('li.rtddt'):
    print li.select('time')[0].text, li.select('h2')[0].text,  \
          li.select('h1')[0].text.strip(),li.select('a')[0]['href']
    ary.append({'time':li.select('time')[0].text, 'category':li.select('h2')[0].text,  \
          'title':li.select('h1')[0].text.strip(),'link':li.select('a')[0]['href']})
```

```python
import pandas as pd
df = pd.DataFrame(ary)
df
```
